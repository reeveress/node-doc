```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib

from matplotlib import pyplot as plt
```

```python
# Constants
c = 3e8
f_c = 12e9
# relative permittivity
e_r = 1
```

```python
f = np.arange(9e9,10e10,100)#,dtype=np.complex)
length = .25
d = .01
k = ((2*np.pi*f)/c)
k_c = np.complex(1.841/(d))
b = np.sqrt(k**2-k_c**2)
A = -20*(np.log10(np.exp(-b*length)))
print(k_c)
print(b)
print(A)
```

```python
# Minimum pipe diameter based on the cutoff frequency
d = (np.arange(1,5,.5))/100
print(d)
fc = (1.841*c)/(2*np.pi*d*np.sqrt(e_r))
print(fc)
d = d*100
```

```python


#plt.subplot(211)
plt.plot(d,(fc/1e9),'-g')
plt.xlabel('Pipe Diameter [cm]')
plt.ylabel('Cutoff Frequency [GHz]')
plt.title('e_r = 1')
plt.show()
```
